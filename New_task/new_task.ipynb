```python
import pandas as pd
import numpy as np
from scipy import stats

```

```python
# Read the downloaded CSV file
df = pd.read_csv('data.csv')
df.head()
```

```python
df2= df.copy()
```

```python
df2.tail(30)
```

```python
# Select Argentina beef consumption
arg_beef = df[(df["country"]=="Argentina") & (df["food_category"]=="beef")]
arg_beef
```

```python
arg_consump = arg_beef["consumption"]
arg_consump
```

```python
arg_consump_mean = arg_consump.mean()
arg_consump_mean
```

```python
rg_beef_samples = np.random.normal(loc=arg_consump_mean, scale=5, size=30)
rg_beef_samples
```

```python
# Select Argentina beef consumption
arg_bag = df[(df["country"]=="Bangladesh") & (df["food_category"]=="beef")]
arg_bag.head()
```

```python
bag_consum = arg_bag["consumption"]
bag_consum
```

```python
bag_consum_mean= bag_consum.mean()
bag_consum_mean
```

```python
bag_beef_samples = np.random.normal(loc=bag_consum_mean, scale=5, size=30)
bag_beef_samples
```

```python
# One-sample t-test
t_stat, p_val = stats.ttest_1samp(bag_beef_samples, 50)
```

```python
print("T-statistic:", t_stat)
```

```python
print("P-value:", p_val)
```
